#How to Resize Images with Python


---

When it comes to assessing a website's page experience, images play a crucial role. In fact, resized and compressed images can improve users' latency on a webpage prompting an increase on the overall engagement metrics. 

Learn how to resize your images so they ultimately fit the viewport of both your site and the SERPs by generating suitable thumbnails. 

```python
from PIL import Image
import os 
import PIL #for manipulating images
import glob
```

```python
#@title Open and Resize an Image at your convenience
image = Image.open('/content/1536007070456.JPEG') #@param {type:"string"}
print(image.size)

resized_image = image.resize((500,500)) #@param {type:"string"}
print(resized_image.size)
resized_image
```

#How to Resize Images in a Fixed Aspect Ratio


---
To protect pictures’ image and width ratio during resizing process, we need to determine a base width for our script.


```python
fixed_height = 420 #Determine a fixed number for creating the ratio.
image = Image.open('/content/1536007070456.JPEG') #Open the Image
height_percent = (fixed_height / float(image.size[1])) #Create the height percent according to your fixed value.
width_size = int((float(image.size[0]) * float(height_percent))) #Create the image width by multiplying it with the proportioned height value.
image = image.resize((width_size, fixed_height), PIL.Image.NEAREST) #Resize and Optimize the image with PIL.Image.NEAREST 
image.save('resized_nearest.jpg') #Save the new image
image #see the output
```

#Create a Thumbnail Image


---
To create a Thumbnail Image we need to set 400 width and 400 height.

To trigger image optimizations, we are going to save the new file with 65% of qwuality enhancement.


```python
#@title Create a Thumbnail Image
image.thumbnail(size=(400,400)) #@param {type:"string"}
image.save('thumbnail_optimized.jpeg', optimize=True, quality=65)
image
```
